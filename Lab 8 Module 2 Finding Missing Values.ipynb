##### Import Necessary Modules:


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.
</p>


The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
# Display basic information about the dataset
print(df.info())

# Display summary statistics of the dataset
print(df.describe())
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Identify missing values for all columns
missing_values = df.isnull().sum()

# Display the missing values for each column
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
# Create a heatmap of missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

# Show the plot
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# Count the number of missing values in the "Employment" column
missing_employment = df['Employment'].isnull().sum()

# Display the result
print(f"Number of missing values in the 'Employment' column: {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Identify the most frequent value in the "Employment" column
most_frequent_employment = df['Employment'].mode()[0]

# Display the result
print(f"The most frequent value in the 'Employment' column is: {most_frequent_employment}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Identify the most frequent value in the "Employment" column
most_frequent_employment = df['Employment'].mode()[0]

# Impute missing values in the "Employment" column with the most frequent value
df['Employment'].fillna(most_frequent_employment, inplace=True)

# Verify if there are any missing values left in the "Employment" column
missing_employment = df['Employment'].isnull().sum()
print(f"Number of missing values in the 'Employment' column after imputation: {missing_employment}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
# Visualize the distribution of the "Employment" column after imputation
plt.figure(figsize=(10, 6))
sns.countplot(x='Employment', data=df, palette='Set2')

# Add labels and title for clarity
plt.title("Distribution of 'Employment' Column After Imputation")
plt.xlabel('Employment Status')
plt.ylabel('Count')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.show()
```
