### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
# Clean the 'ConvertedCompYearly' column to remove any non-numeric or missing values
df_cleaned = df.dropna(subset=['ConvertedCompYearly'])
df_cleaned = df_cleaned[df_cleaned['ConvertedCompYearly'].apply(lambda x: isinstance(x, (int, float)))]

# Plot the distribution (density plot)
plt.figure(figsize=(12, 6))

# Plotting the distribution using Seaborn
sns.kdeplot(df_cleaned['ConvertedCompYearly'], fill=True, color='blue', lw=2)
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)', fontsize=16)
plt.xlabel('Yearly Compensation (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(True)

plt.show()

# Plotting the histogram
plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned['ConvertedCompYearly'], bins=30, kde=True, color='green')
plt.title('Histogram of Yearly Compensation (ConvertedCompYearly)', fontsize=16)
plt.xlabel('Yearly Compensation (USD)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)

plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
# List all column names in the DataFrame
print(df.columns.tolist())

# Filter data for 'Employed, full-time' status
df_full_time = df[df['Employment'] == 'Employed, full-time']

# Calculate the median of the 'ConvertedCompYearly' column for this group
median_compensation = df_full_time['ConvertedCompYearly'].median()

# Output the result
print(f"The median yearly compensation for 'Employed, full-time' respondents is: ${median_compensation:,.2f}")

# Drop rows where 'ConvertedCompYearly' is NaN
df_full_time_cleaned = df_full_time.dropna(subset=['ConvertedCompYearly'])

# Calculate the median
median_compensation = df_full_time_cleaned['ConvertedCompYearly'].median()
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python


# Clean the data by removing rows where 'ConvertedCompYearly' or 'Country' is NaN
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'Country'])

# Set the figure size for better readability
plt.figure(figsize=(12, 8))

# Create the box plot
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_cleaned)

# Rotate the country labels for better visibility
plt.xticks(rotation=90)

# Add labels and a title to the plot
plt.title('Distribution of Yearly Compensation by Country', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Yearly Compensation (USD)', fontsize=12)

# Display the plot
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Step 1: Calculate Q1, Q3, and IQR for 'ConvertedCompYearly'
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define the outlier range using the IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Filter the dataset to remove rows where 'ConvertedCompYearly' is an outlier
df_refined = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Step 4: Check the shape of the original and refined DataFrame
print(f"Original DataFrame shape: {df.shape}")
print(f"Refined DataFrame shape (after removing outliers): {df_refined.shape}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here


# Step 1: Select relevant columns
columns_to_check = ['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']

# Ensure we are working with the refined dataset without outliers
df_refined_corr = df_refined[columns_to_check]

# Step 2: Calculate the correlation matrix
correlation_matrix = df_refined_corr.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))  # Set the size of the plot
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)

# Add title
plt.title('Correlation Heatmap between Compensation, Experience, and Job Satisfaction', fontsize=16)

# Show the plot
plt.show()
```
