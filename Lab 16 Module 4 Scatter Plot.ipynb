#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
import seaborn as sns

# Step 2: Clean the data - remove rows with missing values in 'Age' or 'JobSatPoints_6'
df.dropna(subset=['Age', 'JobSatPoints_6'], inplace=True)

# Step 3: Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='JobSatPoints_6', data=df, palette='viridis')

# Step 4: Customize the plot
plt.title('Scatter Plot of Age vs. Job Satisfaction', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Job Satisfaction', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here

# Step 2: Clean the data - remove rows with missing values in 'ConvertedCompYearly' or 'JobSatPoints_6'
df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'], inplace=True)

# Step 3: Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ConvertedCompYearly', y='JobSatPoints_6', data=df, palette='coolwarm')

# Step 4: Customize the plot
plt.title('Scatter Plot of Compensation vs. Job Satisfaction', fontsize=16)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=14)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
# Step 2: Clean the data - remove rows with missing values in 'Age' or 'JobSatPoints_6'
df.dropna(subset=['Age', 'JobSatPoints_6'], inplace=True)
# Step 2: Clean the data - ensure 'Age' and 'JobSatPoints_6' are numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Step 3: Create the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='JobSatPoints_6', data=df, scatter_kws={'s': 50, 'alpha': 0.6}, line_kws={'color': 'red'}, ci=None)

# Step 4: Customize the plot
plt.title('Scatter Plot of Age vs. Job Satisfaction with Regression Line', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=14)


# Show the plot
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python

# Step 2: Clean the data - ensure 'Age' and 'YearsCodePro' are numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Step 3: Drop rows with NaN values in 'Age' or 'YearsCodePro'
df.dropna(subset=['Age', 'YearsCodePro'], inplace=True)

# Step 4: Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='YearsCodePro', data=df, alpha=0.6)

# Step 5: Customize the plot
plt.title('Scatter Plot of Age vs. Work Experience (YearsCodePro)', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Work Experience (YearsCodePro)', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here


# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Clean the data - Ensure columns are numeric and handle missing values
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Step 3: Drop rows with missing values in relevant columns
df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Age'], inplace=True)

# Step 4: Create the bubble plot
plt.figure(figsize=(12, 8))

# Set bubble size as the 'Age' column (scaled for better visibility)
bubble_size = df['Age'] * 10  # Scale factor for bubble size

# Scatter plot with bubble sizes
sns.scatterplot(x='ConvertedCompYearly', 
                y='JobSatPoints_6', 
                data=df, 
                size=bubble_size, 
                sizes=(50, 1000),  # Adjust range for bubble size
                alpha=0.6, 
                color='b', 
                legend=False)

# Step 5: Customize the plot
plt.title('Bubble Plot: Yearly Compensation vs. Job Satisfaction', fontsize=16)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=14)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=14)
plt.tight_layout()

# Show the plot
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here


# Step 2: Clean the data
# Remove rows with missing values for LanguageHaveWorkedWith and JobSatPoints_6
df.dropna(subset=['LanguageHaveWorkedWith', 'JobSatPoints_6'], inplace=True)

# Step 3: Process 'LanguageHaveWorkedWith'
# We will count the occurrences of each language across respondents
# Split the languages into a list
language_counts = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()

# Step 4: For each language, calculate the average job satisfaction
language_satisfaction = df.explode('LanguageHaveWorkedWith').groupby('LanguageHaveWorkedWith')['JobSatPoints_6'].mean()

# Step 5: Create a dataframe for visualization
languages_df = pd.DataFrame({
    'Language': language_counts.index,
    'Popularity': language_counts.values,
    'AvgJobSatisfaction': language_satisfaction.values
})

# Step 6: Create the scatter plot
plt.figure(figsize=(12, 8))

sns.scatterplot(x='Popularity', 
                y='AvgJobSatisfaction', 
                data=languages_df, 
                s=100, 
                color='b', 
                edgecolor='k', 
                alpha=0.6)

# Step 7: Customize the plot
plt.title('Popularity of Programming Languages vs. Job Satisfaction', fontsize=16)
plt.xlabel('Language Popularity (Number of Respondents)', fontsize=14)
plt.ylabel('Average Job Satisfaction', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python

# Step 1: Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Clean the data (remove rows with missing values in relevant columns)
df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Employment'], inplace=True)

# Step 3: Create the scatter plot
plt.figure(figsize=(12, 8))

# Use a scatter plot with color-coded employment types
sns.scatterplot(x='ConvertedCompYearly', 
                y='JobSatPoints_6', 
                hue='Employment',  # Differentiate by Employment type
                data=df, 
                palette='Set1',  # Use a distinct color palette
                style='Employment',  # Use different markers for different employment types
                s=100,  # Size of points
                alpha=0.6,  # Transparency of points
                markers=['o', 's', 'D', '^', 'v'])  # Different markers for each employment type

# Step 4: Customize the plot
plt.title('Relationship Between Yearly Compensation and Job Satisfaction by Employment Type', fontsize=16)
plt.xlabel('Yearly Compensation (USD)', fontsize=14)
plt.ylabel('Job Satisfaction', fontsize=14)
plt.legend(title='Employment Type', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python


# Step 2: Clean the data (remove rows with missing values in relevant columns)
df.dropna(subset=['YearsCodePro', 'Age', 'Country'], inplace=True)

# Step 3: Create age groups
bins = [20, 30, 40, 50, 60, 70, 80]  # Define age ranges
labels = ['20-30', '31-40', '41-50', '51-60', '61-70', '71-80']  # Labels for the age ranges
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Step 4: Create the scatter plot
plt.figure(figsize=(14, 8))

# Use a scatter plot with color-coded countries and different markers for age groups
sns.scatterplot(x='Age', 
                y='YearsCodePro', 
                hue='Country',  # Differentiate by Country using color
                style='AgeGroup',  # Different markers for different age groups
                data=df, 
                palette='Set1',  # Use a distinct color palette
                markers=['o', 's', 'D', '^', 'v', 'p'],  # Markers for different age groups
                s=100,  # Size of points
                alpha=0.6)  # Transparency of points

# Step 5: Customize the plot
plt.title('Work Experience vs Age Group by Country', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Years of Professional Coding Experience (YearsCodePro)', fontsize=14)
plt.legend(title='Country / Age Group', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
