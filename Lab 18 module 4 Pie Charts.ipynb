### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python

# Step 1: Load the data
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Clean the data (remove rows with missing values in DatabaseWantToWorkWith)
df_cleaned = df.dropna(subset=['DatabaseWantToWorkWith'])

# Step 3: Split the DatabaseWantToWorkWith column by commas
databases = df_cleaned['DatabaseWantToWorkWith'].str.split(',')

# Step 4: Explode the list into individual rows
databases_exploded = databases.explode()

# Step 5: Count the frequency of each database
database_counts = databases_exploded.value_counts().head(5)

# Step 6: Create the pie chart
plt.figure(figsize=(8, 8))
database_counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=90, wedgeprops={'edgecolor': 'black'})

# Step 7: Customize the chart
plt.title('Top 5 Most-Desired Databases to Work With', fontsize=16)
plt.ylabel('')  # Remove the y-label (as it's unnecessary for pie charts)

# Show the plot
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Step 1: Load the data
file_path = "survey-data.csv"
df = pd.read_csv(file_path)

# Step 2: Clean the data (remove rows with missing values in DevType)
df_cleaned = df.dropna(subset=['DevType'])

# Step 3: Split the DevType column by commas (if there are multiple roles per respondent)
devtypes = df_cleaned['DevType'].str.split(',')

# Step 4: Explode the list into individual rows
devtypes_exploded = devtypes.explode()

# Step 5: Count the frequency of each developer role
devtype_counts = devtypes_exploded.value_counts().head(5)

# Step 6: Create the pie chart
plt.figure(figsize=(8, 8))
devtype_counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=90, wedgeprops={'edgecolor': 'black'})

# Step 7: Customize the chart
plt.title('Top 5 Developer Roles', fontsize=16)
plt.ylabel('')  # Remove the y-label (as it's unnecessary for pie charts)

# Show the plot
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here


# Step 1: Clean the data - remove rows with missing values in LanguageHaveWorkedWith
df_cleaned = df.dropna(subset=['LanguageHaveWorkedWith'])

# Step 2: Split the LanguageHaveWorkedWith column into individual languages (if multiple languages are separated by commas)
language_series = df_cleaned['LanguageHaveWorkedWith'].str.split(',', expand=True).stack()

# Step 3: Count the occurrences of each language
language_counts = language_series.value_counts()

# Step 4: Select the top 5 most common languages
top_languages = language_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
top_languages.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3', figsize=(8, 8))

# Customize the plot
plt.title('Top 5 Programming Languages Worked With', fontsize=16)
plt.ylabel('')  # Hide the y-label (not needed for a pie chart)
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here

# Step 1: Clean the data - remove rows with missing values in NEWCollabToolsHaveWorkedWith
df_cleaned = df.dropna(subset=['NEWCollabToolsHaveWorkedWith'])

# Step 2: Split the NEWCollabToolsHaveWorkedWith column into individual tools
tools_series = df_cleaned['NEWCollabToolsHaveWorkedWith'].str.split(',', expand=True).stack()

# Step 3: Count the occurrences of each collaboration tool
tool_counts = tools_series.value_counts()

# Step 4: Select the top collaboration tools (e.g., top 5)
top_tools = tool_counts.head(5)

# Step 5: Plot the bar chart
plt.figure(figsize=(10, 6))
top_tools.plot(kind='bar', color='skyblue')

# Customize the plot
plt.title('Top Collaboration Tools Used by Respondents', fontsize=16)
plt.xlabel('Collaboration Tools', fontsize=14)
plt.ylabel('Number of Respondents', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here

# Step 1: Clean the data - remove rows with missing values in LanguageAdmired
df_cleaned = df.dropna(subset=['LanguageAdmired'])

# Step 2: Split the LanguageAdmired column into individual languages
admired_languages_series = df_cleaned['LanguageAdmired'].str.split(',', expand=True).stack()

# Step 3: Count the occurrences of each admired language
admired_language_counts = admired_languages_series.value_counts()

# Step 4: Select the top 5 most admired languages
top_5_admired_languages = admired_language_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
top_5_admired_languages.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffb3e6','#ff6666','#ffcc99'])

# Customize the plot
plt.title('Top 5 Most Admired Programming Languages', fontsize=16)
plt.ylabel('')  # Remove the y-label as it's not needed for pie chart
plt.tight_layout()

# Show the plot
plt.show()
```
